### HW: 1. Combining the vertical and horizontal filtered images from Sobel filtering. 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv.imread('Sudoku.jpg', cv.IMREAD_GRAYSCALE)


sobel_x = cv.Sobel(img, cv.CV_64F, 1, 0, 3)
sobel_y = cv.Sobel(img, cv.CV_64F, 0, 1, 3)

sobel_comb = cv.magnitude(sobel_x, sobel_y)
sobel_x = cv.convertScaleAbs(sobel_x)
sobel_y = cv.convertScaleAbs(sobel_y)

comb_result = cv.convertScaleAbs(sobel_comb)


plt.figure(figsize=(15, 15))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Vertical Sobel Filter')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Horizontal Sobel Filter')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(comb_result, cmap='gray')
plt.title('Combined Sobal Filter')
plt.axis('off')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Sudoku.jpg', cv.IMREAD_GRAYSCALE)

# sobel_x = cv.Sobel(img, -1, 1, 0, 3)
# sobel_y = cv.Sobel(img, -1, 0, 1, 3)

# sobel_comb = cv.magnitude(sobel_x, sobel_y)
# combine_sobal = cv.convertScaleAbs(sobel_comb)

sobel_x = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=3)
sobel_y = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=3)

# Take the absolute values of the gradients
abs_sobel_x = np.absolute(sobel_x)
abs_sobel_y = np.absolute(sobel_y)

# Combine the two gradients
combine_sobel = cv.addWeighted(abs_sobel_x, 0.5, abs_sobel_y, 0.5, 0)

# Convert the result to uint8 for display
combine_sobel = cv.convertScaleAbs(combine_sobel)

plt.figure(figsize=(15,15))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')


plt.subplot(1, 3, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('vertical sobel filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('horizontal sobel filter')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(combine_sobel, cmap='gray')
plt.title('Combined Magnitude')
plt.axis('off')

plt.show()
```

### 1. Bitwise_or function for combining the filtered images? Give proper explanation for your answer.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('Sudoku.jpg', cv.IMREAD_GRAYSCALE)

sobel_x = cv.Sobel(img, -1, 1, 0, 3)
sobel_y = cv.Sobel(img, -1, 0, 1, 3)


sobel_x = cv.convertScaleAbs(sobel_x)
sobel_y = cv.convertScaleAbs(sobel_y)

# bitwise_or
combine_sobel = cv.bitwise_or(sobel_x, sobel_y)


plt.figure(figsize=(10, 10))
plt.imshow(combine_sobel, cmap='gray')
plt.title('Combined Sobel using bitwise_or')
plt.axis('off')
plt.show()
```
