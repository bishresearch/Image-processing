```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

## Task 1: Enhance image quality to deal with varying lighting conditions.

```python
image = cv2.imread('Train1006.jpg', cv2.IMREAD_GRAYSCALE)

# Apply GHE
ghe_image = cv2.equalizeHist(image)

# Adaptive Histogram Equalization methods
ahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(8, 8))  # tileGridSize (4, 4), (8, 8), (16, 16)
ahe_image = ahe.apply(image)

# CLAHE Methods
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(4, 4))    # clipLimit 3.0, 4.0, 5.0
clahe_image = clahe.apply(image)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(ahe_image, cmap='gray')
plt.title('Adaptive Histogram Equalization (AHE)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(clahe_image, cmap='gray')
plt.title('Contrast Limited AHE (CLAHE)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2 as cv

# Load the image
image_path = "Train1019.jpg"
image = cv.imread(image_path)

# Convert to grayscale
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Adaptive Histogram Equalization
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(4, 4))
enhanced_image = clahe.apply(gray)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(enhanced_image, cmap='gray')
plt.title('Enhanced with CLAHE')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Task 2:Detect edges of vehicles and road markings to assess traffic patterns.

```python
import cv2 as cv

# Load the image
image_path = "Train1006.jpg"
image = cv.imread(image_path)

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# apply Gaussian Blur
blurred = cv.GaussianBlur(gray, (5, 5), 0)

# Canny Edge Detection
edges = cv.Canny(blurred, threshold1=50, threshold2=130)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection (Canny)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Load the image
image_path = "Train1019.jpg"
image = cv.imread(image_path)

gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# using Gaussian Blur 
blurred = cv.GaussianBlur(gray, (5, 5), 0)

# Canny Edge Detection
edges = cv.Canny(blurred, threshold1=50, threshold2=150)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection (Canny)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Task 3: Apply filters to reduce noise and improve the clarity of moving objects

```python
# Load the image
image_path = "Car100.jpg"
image = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(gray_image, (5, 5), 1.4)

# Bilateral Filter
#bilateral_filter = cv2.bilateralFilter(gray_image, d=9, sigmaColor=75, sigmaSpace=75)
bilateral_filter = cv2.bilateralFilter(gray_image, d=5, sigmaColor=75, sigmaSpace=75)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)  
plt.title('Original Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2) 
plt.title('Gaussian Blur')
plt.imshow(gaussian_blur, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)  
plt.title('Bilateral Filter')
plt.imshow(bilateral_filter, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
